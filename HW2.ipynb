```python
import math
```

```python
cur_x = 8 # The algorithm starts at x=8
rate = 0.01 # step size multiplier
precision = 0.00002
# value bigger than initial precision to start
previous_step_size = 1

df = lambda x: (x-1) * math.e**-x
```

```python
counter = 0
while (previous_step_size > precision and counter <= 500) :
    counter+=1
    prev_x = cur_x
    cur_x += -rate * df(prev_x)
    cur_y = 1-cur_x*math.e**-cur_x
    print("%d: x = %f y = %f" % (counter, cur_x, cur_y))
    previous_step_size = abs(cur_x - prev_x)

print("The local minimum occurs at x = %f y= %f" % (cur_x, cur_y))
```
